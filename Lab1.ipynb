```python
#1. Найти минимальное кол-во кластеров, обеспечивающих приемлемый уровень кластеризации.
#Сделаем это с помощью метода локтя

inertia_values = []
for k in range(2, 11):
   kmeans = KMeans( n_clusters = k, n_init=10)
   kmeans.fit( data )
   inertia_values.append( kmeans.inertia_ )

plt.plot(range(2, 11), inertia_values, marker='o')
plt.xlabel("Количество кластеров")
plt.ylabel("Сумма квадратов расстояний")
plt.title("Метод локтя")
plt.show()
#Как мы можем видеть, минимальное количество кластеров с приемлемым уровнем кластеризации - это 3
```

```python
#2. Выполнить кластеризацию данного множества точек методом k-средних по найденному кол-ву кластеров
kmeans = KMeans( n_clusters = 3, n_init=10 )
kmeans.fit( data )
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

#3. Напечатать:
#- количество точек в каждом кластере
#- координаты центроидов кластеров.

print("Метки кластеров:")
print(labels)
print("Координаты центроидов кластеров:")
print(centroids)
labelNum = np.unique(labels)
for v in labelNum:
   print(f"В кластере с меткой {v} содержится {np.sum(labels == v)} точек" )
plt.scatter(data[:,0], data[:,1], c = labels )
plt.scatter(centroids[:,0], centroids[:,1], c = "r")
plt.show()
```

```python
#4. Выполнить кластеризацию (шаг 2) с визуализацией для 3 любых различных значений параметра random_state
# Возьмем random_state = 50

kmeans = KMeans(n_clusters = 3, random_state=50, n_init=10)
kmeans.fit( data )
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

print("Метки кластеров:")
print(labels)
print("Координаты центроидов кластеров:")
print(centroids)
labelNum = np.unique(labels)
for v in labelNum:
   print(f"В кластере с меткой {v} содержится {np.sum(labels == v)} точек" )
plt.scatter(data[:,0], data[:,1], c = labels )
plt.scatter(centroids[:,0], centroids[:,1], c = "r")
plt.show()
```
